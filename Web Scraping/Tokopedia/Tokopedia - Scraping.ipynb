```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
import time


opt = webdriver.ChromeOptions()
opt.add_experimental_option('w3c', False)
driver = webdriver.Chrome(chrome_options=opt)
keyword = input('Masukan keyword : ')
driver.get('https://www.tokopedia.com/')
elem = driver.find_element_by_xpath("//*[@id='zeus-root']/div/div[1]/header/div[2]/form/div/input")#javascript nyari element
elem.send_keys(keyword) #nyari produk apa
elem.submit()
time.sleep(5)
c = 0
for j in range(2,4): #sampe halaman ke 3
    time.sleep(5)
    for i in range(2,65) : #semua produk di hal itu
        try :
            produk = driver.find_element_by_xpath('/html/body/div[1]/div[6]/div/div/div/div/div/div[1]/div[2]/div/div[3]/div[1]/div/div['+str(i)+']/div/a/div[2]/h3')
            harga = driver.find_element_by_xpath('/html/body/div[1]/div[6]/div/div/div/div/div/div[1]/div[2]/div/div[3]/div[1]/div/div['+str(i)+']/div/a/div[2]/div/span/span')
            print(produk.get_attribute('innerHTML') + " --- " + harga.get_attribute('innerHTML'))
        except :
            c = 1
    print("")
    print("================ PAGE "+str(j)+" =================")
    page = "https://www.tokopedia.com/search?st=product&q="+keyword+"&page="+str(j);
    driver.get(page);

# fastrack = WebDriverWait(driver, 10).until(ec.visibility_of_element_located((By.ID, "mitra-tokopedia-click")));
# fastrack.click()
```
