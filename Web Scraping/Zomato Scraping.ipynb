```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import time

driver = webdriver.Chrome()
driver.get("https://www.zomato.com/jakarta/sate-taichan-goreng-tebet/reviews")
time.sleep(5)

count = 1
page = 2
sheet = 2
reviews = []

for i in range(1,3679):
    try:
        review = driver.find_element_by_xpath("/html/body/div[1]/main/div/section[4]/div/div/section[2]/div[2]/p["+str(count)+"]")
        print(str(i)+ ". " + str(review.text))
        reviews.append(review)
        
        if count == 5 :
            time.sleep(3)
            driver.find_element_by_xpath("/html/body/div[1]/main/div/section[4]/div/div/section[2]/div[3]/div[2]/div/a["+str(page)+"]").click()
            count = 0
            time.sleep(3)
            print("================================= Page : " + str(sheet) + " ===========================================")
            sheet += 1
            
            if(sheet == 3):
                page += 2
            else:
                page = 5
            
    except NoSuchElementException:
        print(str(i)+ ". " + "Ga Komen")
        reviews.append("Ga Komen")
        
    count +=1
```
