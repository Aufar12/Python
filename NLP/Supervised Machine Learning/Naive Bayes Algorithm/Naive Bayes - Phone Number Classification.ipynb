```python
import datetime
import numpy as np


jumlahBerurutan = []
awalTanggal = []
totalBerurutan = []

for i in x['A_NUMBER']:
    df = juni_juli_agustus[juni_juli_agustus['A_NUMBER'] == i] #ngambil data tiap nomor
    test1 = []
    
    for j in df['CALLING_TIME']:
        datetime_object = datetime.datetime.strptime(j, '%Y-%m-%d %H:%M:%S') #Ngebaca format tanggal
        d = datetime_object.strftime("%d/%m/%Y") #Ngerubah format tanggal jadi hari bulan tahun
        test1.append(d)

    test2 = sorted(np.unique(test1), key=lambda x: datetime.datetime.strptime(x, "%d/%m/%Y").strftime("%Y-%m-%d")) # Ngebuat tanggal yg sama jadi 1 (Buang duplikat)
    
    counter = -1
    index = 0
    jB = []
    aT = []
    
    a = True   

    if len(test2) < 5: # Kalo tanggal ga sampe 5 hari berbeda yg di telfon
        a = False

    while a:
        temp = datetime.datetime.strptime(test2[index], "%d/%m/%Y") #ngebaca format tanggal yg udh unik
        end_date = temp + datetime.timedelta(days=1) #nambah tanggal + 1 hari
        endDate = end_date.strftime("%d/%m/%Y") #samain format kyk tanggal unik
        if test2[index+1] == endDate: #ngecek kalo hari index +1 sama ga kyk hari +1
            counter+=1
        else: 
            if(counter >= 5): #Kalo berurutan diatas 5 jumlahnya
                jB.append(counter)
                aT.append(test2[index - counter])
                counter = 0
            else: # Kalo berurutan dibawah 5
                counter = 0

        if index == (len(test2)-2): #Kalo udh mau mencapai index, mencegah index out of range
            if(counter >= 5): #Kalo berurutan diatas 5 jumlahnya
                jB.append(counter)
                aT.append(test2[index - counter])
                counter = 0
            else:
                counter = 0

            a = False #Kalo index dh abis keluar dri loop
        else:
            index += 1
            
    jumlahBerurutan.append(jB)
    awalTanggal.append(aT)
    totalBerurutan.append(len(jB))
    
toTrain['TOTAL_CALLING_TIME'] = totalBerurutan
toTrain['TANGGAL_CALLING_TIME'] = awalTanggal
toTrain['JUMLAH_URUTAN_CALLING_TIME'] = jumlahBerurutan

toTrain
```

```python
label = []

for i in x['A_NUMBER']:
    df = toTrain[toTrain['Nomor'] == i] #ngambil data tiap nomor
    temp = 0
    
    durasi = df.get_value(0, 1, takeable = True) 
    if durasi <= durationAverage: #Kalo dia durasi nelfonnya dibawah rata-rata (non fraud)
        temp += 1
    else:
        temp -= 1
    
    kesamaan = df.get_value(0, 2, takeable = True) 
    if kesamaan > 0: # Kalo dia nelfon ke nomor BNUMBER yg memiliki kesamaan lebih dri 0 (non fraud)
        temp += 1
    else:
        temp -= 1
        
    call_time = df.get_value(0, 3, takeable = True)
    if call_time < 1: #Kalo dia nelfon ga pernah 5 hari berturut turut (non fraud)
        temp += 1
    else:
        temp -= 1
        
    if temp > 0:
        label.append('Not Fraud')
    else:
        label.append('Fraud')
        
toTrain['Label'] = label
toTrain
```

```python
#============RUN=====================

fraudPart = toTrain[toTrain.Label  == "Fraud"]
nonFraudPart = toTrain[toTrain.Label  != "Fraud"]

randomFraudPart = fraudPart.sample(2500)
randomNonFraudPart = nonFraudPart.sample(2500)

trainData = pd.concat([randomFraudPart, randomNonFraudPart], ignore_index=True)
fixTrainData = trainData[['Nomor','Durasi','KESAMAAN_B_NUMBER','TOTAL_CALLING_TIME', 'Label']]
fixTrainData
```

```python
#============RUN=====================

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import BernoulliNB
import numpy as np
from sklearn.metrics import confusion_matrix, precision_score, classification_report, recall_score, f1_score

scen_test = input("Please Input Test Ratio (ex: 0.2, 0.3, 0.25, etc) : ")
#Memasukkan data training pada fungsi klasifikasi Naive Bayes
X_train, X_test, y_train, y_test = train_test_split(fixTrainData[['Durasi','KESAMAAN_B_NUMBER','TOTAL_CALLING_TIME']],
                                                    fixTrainData['Label'], test_size=float(scen_test))
#Membuat fungsi klasifikasi Naive Bayes
clf = BernoulliNB().fit(X_train, y_train)

#Menentukan hasil prediksi dari X_test
y_pred = clf.predict(X_test)

print("Accuracy Score Naive Bayes - Bernoulli : " + str((round((accuracy_score(y_test, y_pred)*100),2))) + "%") #score prediksi
print("")
print("=== " + " Naive Bayes - Bernoulli " +" ===")
print("")    
print(classification_report(y_test, y_pred))
print("")
print(pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True))
print("")
```
