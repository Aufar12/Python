```python
data["lower_text"] = data["Comments"].str.lower()
data
```

```python
import string

PUNCT_TO_REMOVE = string.punctuation

def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

data["remove_punct"] = data["lower_text"].apply(lambda text: remove_punctuation(text))
data
```

```python
# import nltk
# nltk.download('stopwords')

from nltk.corpus import stopwords
", ".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

data["remove_stopwords"] = data["remove_punct"].apply(lambda text: remove_stopwords(text))
data
```

```python
from collections import Counter
cnt = Counter()
for text in data["remove_stopwords"].values:
    for word in text.split():
        cnt[word] += 1
        
cnt.most_common(2000)
```

```python
FREQWORDS = set([w for (w, wc) in cnt.most_common(5)])
def remove_freqwords(text):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

data["remove_freqwords"] = data["remove_stopwords"].apply(lambda text: remove_freqwords(text))
data
```

```python
n_rare_words = 10
RAREWORDS = set([w for (w, wc) in cnt.most_common()[:-n_rare_words-1:-1]])
def remove_rarewords(text):
    """custom function to remove the rare words"""
    return " ".join([word for word in str(text).split() if word not in RAREWORDS])

data["remove_rarewords"] = data["remove_freqwords"].apply(lambda text: remove_rarewords(text))
data
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

data["stem_words"] = data["remove_rarewords"].apply(lambda text: stem_words(text))
data
```

```python
# import nltk
# nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer


lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

data["lemmatize_words"] = data["stem_words"].apply(lambda text: lemmatize_words(text))
data
```

```python
def deEmojify(text):
    return text.encode('CP855', 'ignore').decode('CP855')

temp = data["lemmatize_words"]
emot = []

for i in temp :
    x = deEmojify(i)
    emot.append(x)

data['Clean'] = emot
data.head(len(emot))
data
```

```python
import pandas as pd
data = pd.read_excel(r'C:\Users\LENOVO\Downloads\Telegram Desktop\output.xlsx')
del data['Unnamed: 0']
pos = 0
neg = 0
for i in data['Label']:
    if(i == "Positive"):
        pos += 1
    else:
        neg +=1
    
print(pos)
print(neg)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data['Clean'], data ['Label'], test_size=0.2)

print('Train length:', len(X_train))
print('Test length:', len(X_test))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords

vector = TfidfVectorizer(stop_words=stopwords.words('english')).fit(X_train.values.astype('U'))
X_train_vector = vector.transform(X_train.values.astype('U'))
X_test_vector = vector.transform(X_test.values.astype('U'))
print(len(vector.get_feature_names()))
```

```python
from sklearn.tree import DecisionTreeClassifier
cart = DecisionTreeClassifier()
cart.fit(X_train_vector, Y_train)
Y_pred = cart.predict(X_test_vector)
from sklearn import metrics
print("Accuracy: ",metrics.accuracy_score(Y_test, Y_pred))
print("Precision : ",metrics.precision_score(Y_test, Y_pred, average='macro'))
print("Recall : ",metrics.recall_score(Y_test, Y_pred, average='macro'))
print("F1-Score : ",metrics.f1_score(Y_test, Y_pred, average='macro'))
print(metrics.classification_report(Y_test, Y_pred))
print(metrics.confusion_matrix(Y_test, Y_pred))
pd.crosstab(Y_test, Y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
# print(len(vector.get_feature_names()))
# print(X_train_vector)
```
