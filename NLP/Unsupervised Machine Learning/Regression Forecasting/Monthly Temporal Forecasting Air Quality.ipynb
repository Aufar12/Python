```python
import pandas as pd
df = pd.read_excel(r'C:\Users\Aufar\Documents\edm\kualitasUdara.xlsx')
del df['Unnamed: 0']
df = df.dropna()
df
```

```python
import datetime

year2017 = []
monthlyLabel2017 = []
year2018 = []
monthlyLabel2018 = []
year2019 = []
monthlyLabel2019 = []
indeks = df.index
count = 0

for i in df['tanggal']:
    tanggal = datetime.datetime.strptime(i, "%Y-%m-%d")
    
    if tanggal.year == 2017:
        year2017.append(tanggal.month)
        monthlyLabel2017.append(df.loc[ indeks[count] , 'kategori'])
    elif tanggal.year == 2018:
        year2018.append(tanggal.month)
        monthlyLabel2018.append(df.loc[ indeks[count] , 'kategori'])
    else:
        year2019.append(tanggal.month)  
        monthlyLabel2019.append(df.loc[ indeks[count] , 'kategori'])
        
    count += 1
    

print("Jumlah Label di tahun 2017 :" , len(monthlyLabel2017))
print("Jumlah Label di tahun 2018 :" , len(monthlyLabel2018))
print("Jumlah Label di tahun 2019 :" , len(monthlyLabel2019))
```

```python
count = 0
finalMonth2017 = []
detailFinalMonth2017 = []

for i in range(1,13): #Ada 12 Bulan
    temp = year2017.count(i)
    baik = 0
    sedang = 0
    tidakSehat = 0
    
    for i in range(temp):
        tempLabel = monthlyLabel2017[count]
        if tempLabel == "BAIK":
            baik += 1
        elif tempLabel == "SEDANG":
            sedang += 1
        else:
            tidakSehat += 1
            
        count += 1
        
    totalLabelMonthly = {'BAIK': baik, 'SEDANG': sedang, 'TIDAK_SEHAT': tidakSehat}
    detailFinalMonth2017.append(totalLabelMonthly)
    finalMonth2017.append(max(totalLabelMonthly, key=totalLabelMonthly.get))
    
print("Tahun 2017 :", finalMonth2017)
```

```python
count = 0
finalMonth2018 = []
detailFinalMonth2018 = []

for i in range(1,13): #Ada 12 Bulan
    temp = year2018.count(i)
    baik = 0
    sedang = 0
    tidakSehat = 0
    
    for i in range(temp):
        tempLabel = monthlyLabel2018[count]
        if tempLabel == "BAIK":
            baik += 1
        elif tempLabel == "SEDANG":
            sedang += 1
        else:
            tidakSehat += 1
            
        count += 1
        
    totalLabelMonthly = {'BAIK': baik, 'SEDANG': sedang, 'TIDAK_SEHAT': tidakSehat}
    detailFinalMonth2018.append(totalLabelMonthly)
    finalMonth2018.append(max(totalLabelMonthly, key=totalLabelMonthly.get))
    
print("Tahun 2018 :", finalMonth2018)
```

```python
count = 0
finalMonth2019 = []
detailFinalMonth2019 = []

for i in range(1,13): #Ada 12 Bulan
    temp = year2019.count(i)
    baik = 0
    sedang = 0
    tidakSehat = 0
    
    for i in range(temp):
        tempLabel = monthlyLabel2019[count]
        if tempLabel == "BAIK":
            baik += 1
        elif tempLabel == "SEDANG":
            sedang += 1
        else:
            tidakSehat += 1
            
        count += 1
        
    totalLabelMonthly = {'BAIK': baik, 'SEDANG': sedang, 'TIDAK_SEHAT': tidakSehat}
    detailFinalMonth2019.append(totalLabelMonthly)
    finalMonth2019.append(max(totalLabelMonthly, key=totalLabelMonthly.get))
    
print("Tahun 2019 :", finalMonth2019) 
```

```python
finalMonthAllYear = ["January 2017", "February 2017", "March 2017", "April 2017", "May 2017", "June 2017", "July 2017",
                     "August 2017", "September 2017", "October 2017", "November 2017", "December 2017", "January 2018", 
                     "February 2018", "March 2018", "April 2018", "May 2018", "June 2018", "July 2018", "August 2018", 
                     "September 2018", "October 2018", "November 2018", "December 2018", "January 2019", "February 2019", 
                     "March 2019", "April 2019", "May 2019", "June 2019", "July 2019", "August 2019", "September 2019", 
                     "October 2019", "November 2019", "December 2019"]

labelFinal = finalMonth2017 + finalMonth2018 + finalMonth2019
finalDetailAllMonth = detailFinalMonth2017 + detailFinalMonth2018 + detailFinalMonth2019
df_mayoritas = pd.DataFrame({"Month" : finalMonthAllYear, "Detail" : finalDetailAllMonth, "Label" : labelFinal})
df_mayoritas
```

```python
labeltoNumerik = []
for i in df_mayoritas['Label']:
    if i == 'BAIK':
        labeltoNumerik.append(3)
    elif i == 'SEDANG':
        labeltoNumerik.append(2)
    else:
        labeltoNumerik.append(1)
```

```python
from forecast_x import forecast_x as fx

# time series observation
time_series = labeltoNumerik
freq = 12 # monthly
h = 5 # forecast months ahead

# Creating the forecast object
f = fx.forecast(time_series, freq, h)

# Applying any the model from the package
model = f.best_model()
model

# The model variable would produce a list of three elements:
# - Fitting Values
# - Error
# - Forecast
# 4. Valid Model
model = f.model_multi_seas_mean()
model
```

```python
PredictedMonth = ["January 2020", "February 2020", "March 2020", "April 2020", "May 2020"]
Label = []
for i in model[2]:
    if i >= 2.5 and i <= 3:
        Label.append('BAIK')
    elif i >= 2 and i < 2.5:
        Label.append('SEDANG')
    else:
        Label.append('TIDAK SEHAT')
        
df_prediksi = pd.DataFrame({"Month" : PredictedMonth, "Label" : Label})
df_prediksi
```
